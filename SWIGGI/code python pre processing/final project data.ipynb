```python
def Place_Birth(str1):
    if(str(str1)=='Vienna, Austria' or str(str1)=='Vienna'):
        return(1)
    elif(str(str1)=='Budapest'):
        return(4)
    elif(str(str1)=='Prague, Czech Republic'):
        return(5)
    elif(str(str1)=='Strassnitz'):
        return(6)
    else :
        return('nan')
    
def Place_Death(str1):
    if(str(str1)=='Theresienstadt' or str(str1)=='Theresienstadt, Czech Republic'):
        return(7)
    elif(str(str1)=='Treblinka'):
        return(9)
    elif(str(str1)=='Maly Trostinec, Belarus'):
        return(4)
    elif(str(str1)=='Vienna'):
        return(8)
    elif(str(str1)=='Auschwitz-Birkenau, Oswiecim, Oświęcim County, Lesser Poland Voivodeship, P'):
        return(1)
    elif(str(str1)=='Deported with Transport from Vienna, Austria to Nisko, Lwow, Poland'):
        return(5)
    else:
        return(str1)
    
def Deported_From(str1):
    if(str(str1)=='Vienna'):
        return(1)
    elif(str(str1)=='Frankreich'):
        return(3)
    elif(str(str1)=='Drancy'):
        return(2)
    else:
        return(str(str1))
    
    
def Deported_to(str1):
    if(str(str1)=='Nisko'):
        return(5)
    elif(str(str1)=='Kielce'):
        return(2)
    elif(str(str1)=='Riga'):
        return(6)
    elif(str(str1)=='Theresienstadt'):
        return(7)
    elif(str(str1)=='Auschwitz'):
        return(1)
    elif(str(str1)=='Litzmannstadt'):
        return(3)
    elif(str(str1)=='Maly Trostinec, Belarus'):
        return(4)
    else:
        return(str(str1))
```

```python
for i in range(50):
    array_key=[]
    array_type=[]
    array_value=[]
    #----------------------------------------------------------------------------------------------
    array_key.append('Person_ID')
    array_type.append('int')
    array_value.append(str(i+1)) #Person_ID
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['First name']) != 'nan'):
        array_key.append('First_Name')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['First name'])) #First_Name
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Maiden name']) != 'nan'):
        array_key.append('Maiden_Name')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Maiden name'])) #Maiden_Name
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Last name']) != 'nan'):
        array_key.append('Last_Name')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Last name'])) #Last_Name
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Gender']) != 'nan'):
        array_key.append('Gender')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Gender'])) #Gender
    #----------------------------------------------------------------------------------------------
    if(str(1) != 'nan'):
        array_key.append('Religion_ID')
        array_type.append('int')
        array_value.append(str(1)) #Religion_ID
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Date of birth']) != 'nan'):
        array_key.append('Date_OF_Birth')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Date of birth'])) #Date_OF_Birth
    #----------------------------------------------------------------------------------------------
    if(str(Place_Birth(df_persons.iloc[i]['Place of birth'])) != 'nan'):
        array_key.append('Place_Birth')
        array_type.append('int')
        array_value.append(str(Place_Birth(df_persons.iloc[i]['Place of birth']))) #Place_Birth
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Date of death']) != 'nan'):
        array_key.append('Date_OF_Death')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Date of death'])) #Date_OF_Death
    #----------------------------------------------------------------------------------------------
    if(str(Place_Death(df_persons.iloc[i]['Place of death'])) != 'nan'):
        array_key.append('Place_Death')
        array_type.append('int')
        array_value.append(str(Place_Death(df_persons.iloc[i]['Place of death']))) #Place_Death
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['House nr']) != 'nan'):
        array_key.append('House_Number')
        array_type.append('int')
        array_value.append(str(df_persons.iloc[i]['House nr'])) #House_Number
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Street']) != 'nan'):
        array_key.append('Street_Name')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Street'])) #Street_Name
    #----------------------------------------------------------------------------------------------
    if(str(Deported_From(df_persons.iloc[i]['Deported from'])) != 'nan'):
        array_key.append('Deported_From')
        array_type.append('int')
        array_value.append(str(Deported_From(df_persons.iloc[i]['Deported from']))) #Deported_From
    #----------------------------------------------------------------------------------------------
    if(str(Deported_to(df_persons.iloc[i]['Deportation to'])) != 'nan'):
        array_key.append('Deported_To')
        array_type.append('int')
        array_value.append(str(Deported_to(df_persons.iloc[i]['Deportation to']))) #Deported_To
    #----------------------------------------------------------------------------------------------
    if(str(df_persons.iloc[i]['Date of Deportation']) != 'nan'):
        array_key.append('Date_OF_Deport')
        array_type.append('str')
        array_value.append(str(df_persons.iloc[i]['Date of Deportation'])) #Date_OF_Deport
    #----------------------------------------------------------------------------------------------
    print("INSERT INTO Person(", end='')
    for key in range(len(array_key)):
        print(array_key[key], end='')
        if(key!= len(array_key)-1):
            print(",", end='')
    print(")VALUES(", end='')
    
    for key in range(len(array_value)):
        if(array_type[key]=='str'):
            print("\'", end='')
            print(array_value[key], end='')
            print("\'", end='')
            if(key!= len(array_value)-1):
                print(",", end='')
        else:
            print(array_value[key], end='')
            if(key!= len(array_value)-1):
                print(",", end='')
            
    print(")")
```
