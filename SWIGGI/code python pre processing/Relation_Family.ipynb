```python
def insert_relation (df):
    df.reset_index(drop=True, inplace=True)
    array_id=list(df['ID'])
    
    for i in range(df.shape[0]):
        for j in range(df.shape[0]):
            if(i != j):
                print("INSERT INTO Family_relation (Person_ID1,Person_ID2,Type_relation)VALUES("+str(array_id[i])+","+str(array_id[j])+",\'UNkNOW\')"+"--"+df.iloc[i]['First name']+" "+df.iloc[i]['Last name']+" ("+str(array_id[i])+")"+","+df.iloc[j]['First name']+" "+df.iloc[j]['Last name']+" ("+str(array_id[j])+")")
                
    print()
```

```python
df_Lohde = new_data_frame[new_data_frame['Last name'] == 'Lohde']
df_Lohde['ID'] = [24, 25]
print("-------------6-------------")
insert_relation(df_Lohde)

df_Klein = new_data_frame[new_data_frame['Last name'] == 'Klein']
df_Klein['ID'] = [18,19,20]
print("-------------7-------------")
insert_relation(df_Klein)

df_Bettelheim = new_data_frame[new_data_frame['Last name'] == 'Bettelheim']
df_Bettelheim['ID'] = [5,6]
print("-------------8-------------")
insert_relation(df_Bettelheim)

df_Brammer = new_data_frame[new_data_frame['Last name'] == 'Brammer']
df_Brammer['ID'] = [7,8]
print("-------------9-------------")
insert_relation(df_Brammer)

df_Albrecht = new_data_frame[new_data_frame['Last name'] == 'Albrecht']
df_Albrecht['ID'] = [3,4]
print("-------------10-------------")
insert_relation(df_Albrecht)

df_Reiss = new_data_frame[new_data_frame['Last name'] == 'Reiss']
df_Reiss['ID'] = [45,46]
print("-------------11-------------")
insert_relation(df_Reiss)

df_Stern = new_data_frame[new_data_frame['Last name'] == 'Stern']
df_Stern['ID'] = [34,35]
print("-------------12-------------")
insert_relation(df_Stern)

df_Popper = new_data_frame[new_data_frame['Last name'] == 'Popper']
df_Popper['ID'] = [27,28]
print("-------------13-------------")
insert_relation(df_Popper)

df_Ziffer = new_data_frame[new_data_frame['Last name'] == 'Ziffer']
df_Ziffer['ID'] = [38,39]
print("-------------14-------------")
insert_relation(df_Ziffer)

df_Lindner = new_data_frame[new_data_frame['Last name'] == 'Lindner']
df_Lindner['ID'] = [22,23]
print("-------------15-------------")
insert_relation(df_Lindner)
```
